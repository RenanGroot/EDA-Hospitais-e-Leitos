### Carregando os dados

```python
# Importando bibliotecas
import pandas as pd
import plotly.express as px
import plotly.figure_factory as ff

# Biblioteca com os modulos do sistema operacional (sera utilizado para facilitar a leitura dos arquivos)
import os

print("Bibliotecas carregadas!")
```

```python
# Lista dos arquivos dentro da pasta "Database"
files = os.listdir("Database")

df_list = []

# Leitura de cada arquivo da lista criada anteriormente
for file in files:
    df_list.append(pd.read_csv("Database/" + file, on_bad_lines="skip"))

# Junta os DataFrames lidos em um único DataFrame
df_final = pd.concat(df_list)
```

### Valores Nulos

```python
# Soma dos valores nulos de cada coluna do DataFrame
df_final.isna().sum()
```

Para ter uma visão mais clara sobre essas quantidades, iremos observar as porcentagens de valores nulos para cada coluna.

```python
(df_final.isnull().mean() * 100).round(2)
```

Analisando os valores nulos apresentados para cada coluna, há pelo menos uma linha em cada (possivelmente a mesma) coluna, com um valor nulo. Além disso há colunas com grandes porcentagens de valores nulos, o caso das colunas: "MOTIVO DESABILITACAO", "NO_COMPLEMENTO", "NO_EMAIL", "NO_EMAIL","REM INSERTING into EXPORT_TABLE".

Visto que as colunas "REM INSERTING into EXPORT_TABLE" e "MOTIVO DESABILITACAO" possuem apenas valores nulos, sendo a primeira provavelmente resultante de algum "bug" durante a extração. Iremos retira-las do Dataframe.

```python
# Deletando colunas atribuidas ao DataFrame com 100% de valores nulos
df_final.drop(["MOTIVO DESABILITACAO","REM INSERTING into EXPORT_TABLE"], axis=1, inplace=True)
```

```python
# Selecionando linha da coluna "COMP" com valores nulos
df_final[df_final["COMP"].isna()]
```

Visto que realmente há uma linha inteira com valores nulos, vamos retira-la através do metodo `dropna()`.

```python
# Deleta as linhas que possuem valores nulos da coluna "COMP"
df_final.dropna(inplace=True, subset=["COMP"])
```

```python
df_final.isna().sum()
```

Podemos perceber que ela "propagava" a quantidade de nulos para as outras colunas. Agora iremos lidar com os valores nulos das outras colunas

Vamos analisar a coluna "DESC_NATUREZA_JURIDICA" em conjunto com a coluna "NATUREZA_JURIDICA", uma vez que estão correlacionadas. A primeira traz a "descrição" para os valores da segunda.

```python
# DataFrame contendo as linhas (com algumas colunas selecionadas), em que há valores nulos na coluna "DESC_NATUREZA_JURIDICA"
df_final[["CNES","NATUREZA_JURIDICA","DESC_NATUREZA_JURIDICA"]][df_final["DESC_NATUREZA_JURIDICA"].isna()]
```

Podemos perceber que todos os valores nulos possuem a "NATUREZA_JUDIRICA" com o valor 4014 associado, então buscaremos nos outros dados se há uma descrição especifica para este valor em questão.

```python
# DataFrame contendo as linhas em que na coluna "NATUREZA_JURIDICA" está preenchida com 4014
df_nj_nulos = df_final[["COMP","CNES","NATUREZA_JURIDICA","DESC_NATUREZA_JURIDICA"]][df_final["NATUREZA_JURIDICA"] == 4014]

# Mostrando o DataFrame
df_nj_nulos
```

Através dessa busca, por possiveis outros hospitais com o campo "NATUREZA_JURIDICA" igual a 4014. E chegamos aos mesmos valores, então adotaremos outra estratégia: buscar pelos mesmos hosptais que constam nessa lista, através do "CNES" (que é pode ser usado como identificador), em outros periodos para ver se encontramos os valores faltantes.

```python
# DataFrame auxiliar filtrado pelos "CNES" contidos no DataFrame anterior
df_aux = df_final[["CNES","NATUREZA_JURIDICA","DESC_NATUREZA_JURIDICA"]][df_final["CNES"].isin(df_nj_nulos["CNES"])]

# Faz uma contagem sobre a quantidade de linhas para cada "CNES"
df_aux.value_counts("CNES")
```

Analisando, temos 3 "CNES" diferentes que constam esses valores nulos: 2357615, 2318687 e 5199123.

```python
# DataFrame com todos valores do dataset, com a condição de ter um dos CNES descritos anteriormente
#   e agrupados por contagem de valores
df_final[["CNES","DESC_NATUREZA_JURIDICA"]][df_final["CNES"].isin([2357615,2318687,5199123])].value_counts()
```

Observandos valores anteriores, e assumindo que a natureza juridica dessas instituições se mantiveram, preencheremos os valores nulos com o valor de "HOSPITAL_PRIVADO".

```python
# Preenchendo os valores nulos através do metodo .fillna()
df_final["DESC_NATUREZA_JURIDICA"].fillna("HOSPITAL_PRIVADO", inplace=True)
```
