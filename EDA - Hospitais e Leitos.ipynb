### Valores Nulos

```python
# Soma dos valores nulos de cada coluna do DataFrame
df_final.isna().sum()
```

Para ter uma visão mais clara sobre essas quantidades, iremos observar as porcentagens de valores nulos para cada coluna.

```python
(df_final.isnull().mean() * 100).round(2)
```

Analisando os valores nulos apresentados para cada coluna, há pelo menos uma linha em cada (possivelmente a mesma) coluna, com um valor nulo. Além disso há colunas com grandes porcentagens de valores nulos, o caso das colunas: "MOTIVO DESABILITACAO", "NO_COMPLEMENTO", "NO_EMAIL", "NO_EMAIL","REM INSERTING into EXPORT_TABLE".

Visto que as colunas "REM INSERTING into EXPORT_TABLE" e "MOTIVO DESABILITACAO" possuem apenas valores nulos, sendo a primeira provavelmente resultante de algum "bug" durante a extração. Iremos retira-las do Dataframe.

```python
# Deletando colunas atribuidas ao DataFrame com 100% de valores nulos
df_final.drop(["MOTIVO DESABILITACAO","REM INSERTING into EXPORT_TABLE"], axis=1, inplace=True)
```

```python
# Selecionando linha da coluna "COMP" com valores nulos
df_final[df_final["COMP"].isna()]
```

Visto que realmente há uma linha inteira com valores nulos, vamos retira-la através do metodo `dropna()`.

```python
# Deleta as linhas que possuem valores nulos da coluna "COMP"
df_final.dropna(inplace=True, subset=["COMP"])
```

```python
df_final[df_final.columns[df_final.isnull().any()]].isnull().sum()
```

Podemos perceber que ela "propagava" a quantidade de nulos para as outras colunas. Agora iremos lidar com os valores nulos das outras colunas.

Uma estratégia interessante, que podemos utilizar, é de buscar preencher esses dados nulos baseando-se em valores válidos na mesma coluna, e que possuem o mesmo identificador ("CNES"). Utilizaremos essa estratégia a seguir:

```python
# Preenche os valores nulos, baseando-se no primeiro valor valido encotrado ao se agrupar pelo "CNES"
df_final = df_final.fillna(df_final.groupby(["CNES"]).transform('first'))
```

```python
df_final[df_final.columns[df_final.isnull().any()]].isnull().sum()
```

Agora precisamos lidar com os valores nulos das colunas restantes. Iniciando com as colunas "NU_ENDERECO" e "NO_COMPLEMENTO" que possuiem informações sobre o endereço. A primeira coluna expressa o numero do edificio, que pode estar incluso no campo logradouro dependendo a forma com a qual foi preenchia; enquanto que a segunda, por se tratar de um campo que complementa o endereço, muitas vezes é deixado em branco.

Em busca de mais detalhes sobre esses campos faltantes, iremos selecionar os dados em que estão com o campo "NU_ENDERECO" nulo, e traremos juntos seu "NO_LOGRADOURO".

```python
# DataFrame contendo as linhas (com algumas colunas selecionadas), em que há valores nulos na coluna "NU_ENDERECO"
df_final[["CNES","NO_LOGRADOURO","NU_ENDERECO"]][df_final["NU_ENDERECO"].isna()].drop_duplicates()
```

Realmente há dados em que o numero esta incluso ao logradouro por isso preencheremos os campos nulos, da coluna "NU_ENDERECO" com a mensagem: "vide lougradouro, ou s/n". Enquanto que na coluna "NO_COMPLEMENTO", iremos preencher os campos nulos com "s/complemento".

```python
# Preenchendo os valores nulos através do metodo .fillna()
df_final["NU_ENDERECO"].fillna("vide lougradoura, ou s/n", inplace=True)

df_final["NO_COMPLEMENTO"].fillna("s/complemento", inplace=True)
```

```python
df_final[df_final.columns[df_final.isnull().any()]].isnull().sum()
```

Aproveitando que estamos lidando com campos relacionados a localização do hospital, vamos analisar o valores nulos presentes no "NO_BAIRRO", que indica o bairro do hospital em questão.

```python
# DataFrame contendo as linhas (com algumas colunas selecionadas), em que há valores nulos na coluna "NO_BAIRRO"
df_final[["CNES","NO_LOGRADOURO","NO_BAIRRO","NU_ENDERECO"]][df_final["NO_BAIRRO"].isna()].drop_duplicates()
```

Apesar de constar 3 valores nulos na contagem geral, todos pertencem ao mesmo hospital em questão. A partir disso, vamos tentar encontrar outros registros desse hospital em nosso dataset a fim de completar esses valores nulos.

```python
# DataFrame contendo as linhas (com algumas colunas selecionadas), em que o CNES é igual a 262080
df_final[["COMP","CNES","NO_BAIRRO"]][df_final["CNES"]==262080].drop_duplicates()
```

O bairro faltante no hospital em questão, assumindo que ele não mudou de endereço, é "SAO CRISTOVAO", e preencheremos os valores faltantes com essa informação.

```python
# Preenchendo os valores nulos através do metodo .fillna()
df_final["NO_BAIRRO"].fillna("SAO CRISTOVAO", inplace=True)
```

```python
df_final[df_final.columns[df_final.isnull().any()]].isnull().sum()
```
